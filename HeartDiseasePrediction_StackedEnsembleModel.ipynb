```python
# Data Preprocessing
from sklearn.preprocessing import StandardScaler

# Data Splitting
from sklearn.model_selection import train_test_split

# Hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV

# Base Models
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,roc_curve,classification_report

# Stacking Model
from mlxtend.classifier import StackingCVClassifier
```

### 2. Read the cleaned data.

(5 Points)

```python
heartAttack_data = pd.read_csv("data_heartAttack.csv")
```

### 3. Divide the dataset into train and validation DataFrames.

(5 Points)

```python
X = heartAttack_data.drop('target', axis=1)
y = heartAttack_data["target"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### 4. Construct an Ensemble model (STACKING Ensemble) using 2 base classifier/regressor and 1 stacked model.

`NOTE:`
1.   Hyperparameter Tuning: Choose a base classifier model and define the range of hyperparameter values corresponding to the model to be searched over for hyperparameter tuning. Use RandomizedSearchCV or any Hyperparameter tuning techniques.
2.   Once the tuning is complete, find the position (iteration number) at which the highest mean test score was obtaines. Find the corresponding hyperparameters to be used for your based models and stacked model.
3.   Split the dataset intro training and validation sets and train a new model, using the final hyperparameters on the training dataset. Use scikit-learn's train_test_split() method to split X and y to train and test components, with test comprising 15% of the dataset or any which can attain your target accuracy:
4.   Train both your base models and stacked model using the final hyperparameters used from the hyperparameter tuning done in step 2.



(10 Points)

#### Base Classifier

##### Extreme Gradient Boost

###### Hyperparemeter Tuning for Extreme Gradient Boost

```python
xgb_params = {'learning_rate': [0.01, 0.03, 0.001], 
              'n_estimators': [25, 100, 1000], 
              'max_depth': [6, 10, 15],
              'gamma': [0, 0.6], 
              'subsample':[0.52, 1],
              'colsample_bytree': [0.6, 1],
              'seed': [27], 
              'reg_lambda': [1, 2], 
              'booster': ['gbtree', 'dart', 'gblinear'], 
              'colsample_bylevel': [0.6], 
              'colsample_bynode': [0.5, 0.6]}

xgb_base = XGBClassifier(**xgb_params)
xgb_cv = RandomizedSearchCV(xgb_base, xgb_params, cv= 10)
xgb_cv.fit(X_train, y_train)

print("Tuned Desicion XGB Parameters: {}".format(xgb_cv.best_params_))
print("Best score is {}: ".format(xgb_cv.best_score_))
```

```python
xgb = XGBClassifier(learning_rate=0.01, 
                    n_estimators=25, 
                    max_depth=15,
                    gamma=0.6, 
                    subsample=0.52,
                    colsample_bytree=0.6,
                    seed=27, 
                    reg_lambda=2, 
                    booster='dart', 
                    colsample_bylevel=0.6, 
                    colsample_bynode=0.5)
xgb.fit(X_train, y_train)
xgb_train_prediction = xgb.predict(X_train)
```

##### Support Vector Classifier

###### Hyperparameter Tuning for Support Vector Classifier

```python
svc_params = {'kernel': ('rbf', 'poly'),
              'C': (1.0, 2.0, 3.0, 4.0)}

svc_base = SVC(**svc_params)
svc_cv = RandomizedSearchCV(svc_base, svc_params, cv= 10)
svc_cv.fit(X_train, y_train)

print("Tuned Desicion SVC Parameters: {}".format(svc_cv.best_params_))
print("Best score is {}: ".format(svc_cv.best_score_))
```

```python
svc =  SVC(kernel='rbf', C=2)
svc.fit(X_train, y_train)
svc_train_prediction = svc.predict(X_train)
```

#### Stacked Model

##### Stacking CV Classifier

```python
scv = StackingCVClassifier(classifiers=[xgb, svc],meta_classifier= KNeighborsClassifier(n_neighbors=10), random_state=42)
scv.fit(X_train,y_train)
scv_train_prediction = scv.predict(X_train)
```

### 5. Calculate the performance and use a metric accordingly. <font color=red>Target Score: 80% and Above </font>.

(5 Points)

#### Base Classifier

##### Accuracy Score of Extreme Gradient Boost on Training Data.

```python
xgb_conf_matrix = confusion_matrix(xgb_train_prediction, y_train)
xgb_acc_train_score = accuracy_score(xgb_train_prediction, y_train)
print("Accuracy of Extreme Gradient Boost:",xgb_acc_train_score * 100,'\n')
print(classification_report(xgb_train_prediction, y_train))
```

##### Accuracy Score of Support Vector Classifier on Training Data.

```python
svc_conf_matrix = confusion_matrix(svc_train_prediction, y_train)
svc_acc_train_score = accuracy_score(svc_train_prediction, y_train)
print("Accuracy of Support Vector Classifier:",svc_acc_train_score * 100,'\n')
print(classification_report(svc_train_prediction, y_train))
```

#### Stacked Model

##### Accuracy Score of StackingCVClassifier on Training Data.

```python
scv_conf_matrix = confusion_matrix(scv_train_prediction, y_train)
scv_acc_train_score = accuracy_score(scv_train_prediction, y_train)
print("Accuracy of StackingCVClassifier:",scv_acc_train_score * 100,'\n')
print(classification_report(scv_train_prediction, y_train))
```
